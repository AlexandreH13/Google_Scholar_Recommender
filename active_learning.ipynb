```python
import pandas as pd
```

```python
df1 = pd.read_csv("raw_data_com_labels.csv", index_col=0)
df1 = df1[df1['y'].notnull()]
df1.head()
```

```python
df2 = pd.read_csv("data_active_learning_com_labels.csv", index_col=0)
df2['novo'] = 1
df2.shape
```

```python
df = pd.concat([df1, df2.drop("p", axis=1)])
```

```python
df.head()
```

```python
df.shape
```

```python
df_limpo = df.copy()
```

```python
df_limpo['novo'] = df['novo'].fillna(0)
df_limpo.head()
```

```python
df_limpo = df_limpo.reset_index(drop=True)
df_limpo.head()
```

```python
def clean_data(cols_list, char_list):
    for col in cols_list:
        for char in char_list:
            df_limpo[col] = df_limpo[col].str.replace(char, '')
```

```python
cols_to_clean = ['citacoes', 'ano', 'tipo_arquivo', 'versao']
spec_chars = ["[", "]", "'", " "]

clean_data(cols_to_clean, spec_chars)
df_limpo.head()
```

```python
cols_to_convert = ['citacoes', 'versao']
df_limpo[cols_to_convert] = df_limpo[cols_to_convert].astype('int64')
```

```python
df_limpo.dtypes
```

```python
for i in df_limpo['ano']:
    if(len(i)>4):
        print(i)
```

```python
coluna_ano = df_limpo['ano'].copy()
```

```python
import re

for i in range(len(coluna_ano)):
    print(i)
    if(len(coluna_ano[i])>4):
        res = re.findall(r'\d{4}', coluna_ano[i])
        if(len(res)>1):
            coluna_ano[i] = res[len(res)-1]
        else:
            coluna_ano[i] = res[0]
print(coluna_ano)
```

```python
df_limpo['ano'] = coluna_ano
df_limpo.head()
```

```python
df_limpo['ano'].astype('int32')
```

```python
from sklearn.preprocessing import OneHotEncoder
encoded_columns = df_limpo[['tipo_arquivo', 'busca']]
encoder = OneHotEncoder(handle_unknown='ignore')

enc_columns = pd.DataFrame(encoder.fit_transform(encoded_columns).toarray())
enc_columns.head()
```

```python
new_labels = ['CITACAO', 'DOC', 'HTML', 'LIVRO', 'PDF', 'algoritmos_geneticos', 'machine_learning']
enc_columns.columns = new_labels
```

```python
df_limpo = df_limpo.join(enc_columns)
df_limpo.head()
```

```python
df_limpo = df_limpo.drop(columns=['tipo_arquivo', 'busca'])
df_limpo.head()
```

#### Transformando a variável 'titulo'

```python
coluna_titulo = df_limpo['titulo']
coluna_titulo
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(min_df=2)
titulo_vec = vectorizer.fit_transform(coluna_titulo)
```

```python
titulo_vec.shape
```

```python
df_titulo = pd.DataFrame(titulo_vec.toarray(), columns=vectorizer.get_feature_names())
df_titulo.head()
```

```python
df_limpo = df_limpo.join(df_titulo)
df_limpo.head()
```

### Agora, com mais dados de treino, verificamos o resultado do modelo

```python
X = df_limpo.drop(['y', 'titulo', 'link'], axis=1)
y = df_limpo.y
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=1000, random_state=0, class_weight="balanced", n_jobs=2)
rf.fit(X_train, y_train)
```

```python
predict = rf.predict_proba(X_test)[:, 1]
predict
```

```python
from sklearn.metrics import roc_auc_score, average_precision_score

auc = roc_auc_score(y_test, predict)
ap = average_precision_score(y_test, predict)

print("AUC: {} | AP: {}".format(auc, ap))
```
